# Prime Number Experiments
## Is it Prime?

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True
```

```python
print([i for i in range(1, 50) if is_prime(i)]) # list comprehesion
```

## Primes up to n

```python
def primes_up_to(n):
    return [i for i in range(2, n+1) if is_prime(i)]
```

```python
primes_up_to(20)
```

## Visualising Prime Gaps

```python
import matplotlib.pyplot as plt

primes = primes_up_to(100)
gaps = [primes[i+1] - primes[i] for i in range(len(primes)-1)]

plt.plot(gaps)
plt.title("Prime Gaps")
plt.xlabel("Index")
plt.ylabel("Gap Size")
plt.grid(True)
plt.show()
```

## Prime Factorisation

```python
def prime_factors(n):
    factors = []
    i = 2
    while i * i <= n:
        while n % i == 0:
            factors.append(i)
            n //= i
        i += 1
    if n > 1:
        factors.append(n)
    return factors
```

```python
prime_factors(20)
```
